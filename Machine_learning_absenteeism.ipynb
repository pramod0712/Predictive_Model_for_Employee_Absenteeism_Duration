```python
import pandas as pd
import numpy as np
```

```python
data=pd.read_csv('Absenteeism_preprocessed.csv')
```

```python
data.head()
```

```python
data.describe()
```

```python
targets=np.where(data['Absenteeism Time in Hours']<=data['Absenteeism Time in Hours'].median(),0,1)
targets
```

```python
data['Excessive Absenteeism']=targets
```

```python
data=data.drop('Absenteeism Time in Hours',axis=1)
```

```python
data
```

```python
inputs=data.iloc[:,:-1]
inputs
```

```python
targets=data.iloc[:,-1]
targets
```

```python
from sklearn import preprocessing
from sklearn.model_selection import train_test_split
```

```python
inputs=preprocessing.scale(inputs)
```

```python
inputs
```

```python
inputs.shape
```

# Splitting the data

```python
x_train,x_test,y_train,y_test=train_test_split(inputs,targets,test_size=0.2,random_state=69)
```

```python
from sklearn.linear_model import LogisticRegression 
from sklearn.metrics import accuracy_score
```

```python
reg=LogisticRegression()
```

```python
reg.fit(x_train,y_train)
```

```python
reg.score(x_train,y_train)
```

```python
prediction=reg.predict(x_train)
```

```python
x_train is prediction
```

```python
accuracy_score(prediction,y_train)
```

```python
features=data.iloc[:,:-1].columns.values
features
```

```python
summary=pd.DataFrame(columns=['Feature Name'],data=features)
```

```python
coef=reg.coef_
```

```python
reg.intercept_
```

```python
summary['Weights']=np.transpose(coef)
```

```python
summary
```

```python
reg.score(x_test,y_test)
```
